```python
import pandas as pd
```

```python
data=pd.read_csv('CarPrice_Assignment.csv',sep=',')
```

```python
print(data)
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
b=['car_ID', 'symboling', 'CarName', 'fueltype', 'aspiration',
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
       'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg',
       'price']
for i in b:
        data[i]=a.fit_transform(data[i])
data
```

```python
x=data.drop(['price','car_ID'],axis=1)
print(x)
```

```python
#'carwidth', 'carheight', 'curbweight', 'enginetype
data[['carwidth','carheight','curbweight','enginetype']]
```

```python
y=data['price']
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
reg1=LinearRegression()
reg1.fit(xtrain,ytrain)
```

```python
accuracy=reg1.score(xtest,ytest)*100
print(accuracy)
```

```python
predict=reg1.score(xtrain,ytrain)*100
print(predict)
```

```python
v=[[5,2,1,0,1,0,2,0,2,22,8,1,92,0,2,18,5,23,4,13,30,10,8,10]]
out=reg1.predict(v)
print(out)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg2=DecisionTreeRegressor()
reg2.fit(xtrain,ytrain)
```

```python
accuracy=reg2.score(xtest,ytest)*100
print(accuracy)
```
